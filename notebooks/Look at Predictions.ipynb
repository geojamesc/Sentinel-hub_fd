```python
from eolearn.core import EOPatch
import matplotlib.pyplot as plt
```

```python
eop = EOPatch.load('/home/james/Work/FieldBoundaries/Sentinel-hub_fd/input-data/eopatches/32VNH_1')
```

```python
print(eop.data.keys())
```

```python
print(len(eop.data['EXTENT_PREDICTED_folds_avg_10e']))
```

```python
tidx = 0 # select one timestamp
viz_factor = 2.5

fig, axs = plt.subplots(figsize=(15, 5), ncols=3, sharey=True)
axs[0].imshow(viz_factor * eop.data['BANDS'][tidx][..., [2,1,0]]/10000)
axs[0].set_title('RGB bands')
#axs[1].imshow(eop.data['EXTENT_PREDICTED_folds_avg_10e'][tidx].squeeze(), vmin=0, vmax=1)
axs[1].imshow(eop.data['EXTENT_PREDICTED_folds_avg_10e'][tidx].squeeze())
axs[1].set_title('Extent')
axs[2].imshow(eop.data['BOUNDARY_PREDICTED_folds_avg_10e'][tidx].squeeze(), vmin=0, vmax=1)
axs[2].set_title('Boundary');
```

```python
tidx = 0  # select one timestamp 
viz_factor = 3.5

fig, axs = plt.subplots(figsize=(15, 5), ncols=4, sharey=True)
axs[0].imshow(viz_factor * eop.data['BANDS'][tidx][:200, :200, [2,1,0]]/10000)
axs[0].set_title('RGB bands')
#axs[1].imshow(eop.data['EXTENT_PREDICTED_folds_avg_10e'][tidx].squeeze()[:500, :500], vmin=0, vmax=1, alpha=.2)
axs[1].imshow(eop.data['EXTENT_PREDICTED_folds_avg_10e'][tidx].squeeze()[:200, :200])
axs[1].set_title('Extent')
#axs[2].imshow(eop.data['BOUNDARY_PREDICTED_folds_avg_10e'][tidx].squeeze()[:500, :500], vmin=0, vmax=1, alpha=.2)
axs[2].imshow(eop.data['BOUNDARY_PREDICTED_folds_avg_10e'][tidx].squeeze()[:200, :200])
axs[2].set_title('Boundary')
#axs[3].imshow(eop.data['DISTANCE_PREDICTED_folds_avg_10e'][tidx].squeeze()[:500, :500], vmin=0, vmax=1, alpha=.2)
axs[3].imshow(eop.data['DISTANCE_PREDICTED_folds_avg_10e'][tidx].squeeze()[:200, :200])
axs[3].set_title('Distance')
```
